#**Ecuación Maestra.**

```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import*
%pylab inline
```

##**Descripción cuántica.**

En la aproximación de onda rotante (RWA), el Hamiltoniano mecánico cuántico  $H$ que describe la dinámica coherente del acople cavidad-punto cuántico es dado por:

   $$H=\omega_{a}\sigma^{\dagger}\sigma+\omega_{c}a^{\dagger}a+ig(a\sigma^{\dagger}-a^{\dagger}\sigma).$$
   
Aquí $\omega_{c}$ y $\omega_{a}$ son las frecuencias respectivas de la cavidad y el punto cuántico. $\sigma$ es el operador escalera (de bajada) del QD; $a$ es el operador de aniquilación para un modo de la cavidad. El Hamiltoniano de Jaynes-Cummings rotado en la frecuencia $\omega_{l}$ es

$$\textbf{H}=\Delta_{a}\sigma^{\dagger}\sigma+\Delta_{c}a^{\dagger}a+ig(a\sigma^{\dagger}-a^{\dagger}\sigma)+i\Omega(t)(a-a^{\dagger}).$$

##**Parametros del problema.**

```python
wc = 1*2*np.pi    # cavity frequency
wa = 2*2*np.pi    # atom frequency
wl = 3*2*np.pi   # laser frequency
g  = 25e9*2*np.pi  # coupling strength
kappa = 29e9*2*np.pi # cavity dissipation rate
gamma = 1e9*2*np.pi   # atom dissipation rate
omega= 1e9*2*np.pi #peak intensity of the driving pulse
N = 15           # number of cavity fock states

use_rwa = True
```

```python
tlist=np.linspace(0,60e-12,100) # in ps
```

##**Estado inicial del Hamiltoniano. **

```python
# intial state
psi0 = tensor(basis(N,0), basis(2,1))    # start with an excited atom

# operators
a  = tensor(destroy(N), qeye(2))
sm = tensor(qeye(N), destroy(2))

# Hamiltonian
if use_rwa:
    H0 = (wc-wl)* a.dag() * a + (wa-wl)* sm.dag() * sm + 1j*g * (a.dag() * sm - a * sm.dag())
else:
    H0 = (wc-wl)* a.dag() * a + (wa-wl)* sm.dag() * sm + 1j*g * (a-a.dag()) * (sm + sm.dag())

H1=1j*omega*(a-a.dag())
```


##**Evolución del sistema.**

La evolución del sistema se da solucionando los términos de Lindbland de la ecuación maestra, requerimos los valores esperados de $a^{\dagger}a$ y $σ^{\dagger}σ$ son solucionados al pasar por la lista [a.dag()*a, sm.dag()*sm] en el quinto argumento del solucionador.


```python
for p in tlist:

    Ha = H0 +np.exp(-(1/2)*((p-10e-12)/(1e-12))**2) * H1
    

    output1 = mesolve(Ha, psi0, tlist, [sqrt(kappa) * a, sqrt(gamma) * sm.dag()*sm], [a.dag() * a])
    
    n_c1 = output1.expect[0]
    plt.plot(tlist, 37.5*n_c1,color="red",linewidth=2.5,linestyle="-",label="bombeo" )
    #plt.plot(tlist,37*n_c2,color="blue",linewidth=2.5,linestyle="-",label="sin bombeo" )
    plt.plot(tlist,16*np.exp(-(1/2)*((tlist-10e-12)/(1e-12))**2),color="green",linewidth=2.5,linestyle="dashed",label="laser")
```

##**Resultados.**

```python
output2=mesolve(H0, psi0, tlist, [sqrt(kappa) * a, sqrt(gamma) *sm], [a.dag() * a])
n_c2 = output2.expect[0]

plt.plot(tlist, n_c1,color="red",linewidth=2.5,linestyle="-",label="bombeo" )
plt.plot(tlist, n_c2,color="blue",linewidth=2.5,linestyle="-",label="sin bombeo" )
plt.plot(tlist,np.exp(-(1/2)*((tlist-10e-12)/(1e-12))**2),color="green",linewidth=2.5,linestyle="dashed",label="laser")
plt.xlabel("Time(ps)",fontsize=16)
plt.ylabel("Cavitiy Transmission $(a{a}^{\dagger})$",fontsize=16)
plt.title("Vacuum Rabi Oscillations", fontsize=16)

plt.legend(loc='best')
```
